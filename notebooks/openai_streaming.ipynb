```python
from pydantic import BaseModel

# from farmwise.schema import MessageReady
from openai.types.responses import ResponseTextDeltaEvent
from agents import Agent, Runner, RawResponsesStreamEvent

from typing import AsyncIterator, Any


class MessageReady(BaseModel):
    """
    Emitted whenever a complete line (i.e. content up to the next newline)
    of assistant text is available.
    """

    content: str  # the plain-text line (no JSON cruft)
    type: str = "message_ready"  # fits the naming style of other events


# --------------------------------------------------------------------------- #
#  Your new event                                                             #
# --------------------------------------------------------------------------- #

# Alternative approach if you want to process content character by character as it streams
async def process_stream_events(event_stream: AsyncIterator[Any]) -> AsyncIterator[Any]:
    accumulated = ""
    start_token = '"content":"'
    end_token = '","actions":'
    message_ready_token = "\\n\\n"
    in_content = False

    async for event in event_stream:
        if event.type == RawResponsesStreamEvent.type and isinstance(event.data, ResponseTextDeltaEvent):

            delta = event.data.delta
            accumulated += delta

            if start_token in accumulated:
                in_content = True
                _, accumulated = accumulated.split(start_token)

            if in_content:
                if message_ready_token in accumulated:
                    ready, accumulated = accumulated.split(message_ready_token)
                    yield MessageReady(content=ready)

            if end_token in accumulated:
                ready, accumulated = accumulated.split(end_token)
                yield MessageReady(content=ready)
                in_content = False

        # Pass through original events
        yield event


async def main():
    agent = Agent(
        name="Agronomy expert",
        instructions="You are an expert in agronomy",
        output_type=WhatsAppResponse
    )

    result = Runner.run_streamed(
        agent,
        input="Tell me about sustainable land management practices in Kenya.",
    )

    async def async_iter_from_list(_events):
        for item in _events:
            yield item

    # async for event in process_stream_events(async_iter_from_list(events)):
    async for event in process_stream_events(result.stream_events()):
        if isinstance(event, MessageReady):
            # do whatever you need with the completed line
            print("READY:", event.content)
        elif not isinstance(event, RawResponsesStreamEvent):
            print(event)


await main()
```
